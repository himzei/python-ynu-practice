```python
# csv 파일 쓰기

import csv 

with open("./example.csv", "w", encoding="utf8") as file: 

    csv_writer = csv.writer(file)
    print(type(csv_writer))

    csv_writer.writerow(["이름", "나이", "직업"])

    for i in range(10): 
        csv_writer.writerow([f"{i}", "홍길동", "30", "학생"])
        
    csv_writer.writerow(["철수", "20", "교사"])
    csv_writer.writerow(["영희", "25", "경찰"])
```

```python
# csv 파일 읽기

import csv

with open("./FY2019.csv", "r") as file:

    csv_reader = csv.reader(file)

    for row in csv_reader: 
        # print(row[1], row[8])
        print(row[0:4])
        
```
